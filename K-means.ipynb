```python
# Seleccionar las características para entrenar el modelo
X = df.iloc[:, 2:].values  # Ignoramos la columna 'CustomerID' y 'Gender'

# Mostrar las primeras filas de las características seleccionadas
print("Características seleccionadas:\n", X[:5])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Lista para almacenar los valores de inercia
inertia = []

# Rango de número de clústeres
k_range = range(2, 11)

# Aplicar el algoritmo K-means para cada valor de k
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Graficar la curva de codo
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Número de clústeres')
plt.ylabel('Inercia')
plt.title('Método del codo')
plt.show()

# Calcular el valor de la silueta para cada número de clústeres
silhouette_scores = [silhouette_score(X, KMeans(n_clusters=k, random_state=42).fit_predict(X)) for k in k_range]

# Graficar el valor de la silueta
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Número de clústeres')
plt.ylabel('Valor de la silueta')
plt.title('Método de la silueta')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score

# Calcular el coeficiente de Silhouette
silhouette_score_kmeans = silhouette_score(X, kmeans.labels_)
print("Coeficiente de Silhouette para K-means:", silhouette_score_kmeans)

# Calcular el índice de Calinski-Harabasz
calinski_harabasz_score_kmeans = calinski_harabasz_score(X, kmeans.labels_)
print("Índice de Calinski-Harabasz para K-means:", calinski_harabasz_score_kmeans)
```

```python
import matplotlib.pyplot as plt

# Ajustar el modelo k-means al conjunto de datos
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualización de los clústeres
plt.figure(figsize=(10, 6))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroides')
plt.title('Clústeres de clientes')
plt.xlabel('Ingresos anuales (k$)')
plt.ylabel('Puntuación de gasto (1-100)')
plt.legend()
plt.show()
```
